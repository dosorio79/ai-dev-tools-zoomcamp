```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('../')
import search
from minsearch import Index
```

```python
# PARAMS
fast_mcp_repo_url = "https://github.com/jlowin/fastmcp/archive/refs/heads/master.zip"
zip_path = "../data/index"
zip_file = "mcp_index.zip"
```

```python
# Download the zip file
search.download_zip_file_from_url(fast_mcp_repo_url, zip_path, zip_file)
```

```python
# get the zip root folder
zip_root = search.discover_zip_root(zip_path, zip_file)
print(f"Zip root: {zip_root}")
```

```python
# Parse markdown files from the zip file
md_files = search.parse_zip_file( "../data/index", "mcp_index.zip")
md_files
```

```python
# Create docs from parsed data
docs = search.create_docs_from_zip_file(zip_path, zip_file, zip_root, md_files)
```

```python
index = search.create_search_index(docs)
```
